```python
import math
def read_path_from_file(file_path):
    path = []  # 存储路径的列表
    # time=0
    # 打开文件并逐行读取路径
    with open(file_path, 'r') as file:
        for line in file:
            # 按空格分割每一行，得到时间点的位置坐标
            coordinates = line.strip().split()
            # 将位置坐标转换为浮点数，并添加到路径列表中
            path.append((float(coordinates[0]), float(coordinates[1])))
            # time+=1
            # if time==100:
            #     break;  
    
    return path
def calculate_users_covered(users, radius, position):
    # 初始化覆盖范围内用户数量
    users_covered = 0
    
    # 遍历所有用户
    for user in users:
        # 计算用户与位置之间的距离
        distance = math.sqrt((user[0] - position[0])**2 + (user[1] - position[1])**2)
        
        # 如果距离小于等于覆盖范围半径，则该用户在覆盖范围内
        if distance <= radius:
            users_covered += 1
    
    return users_covered
def total_business_volume(path, users, radius):
    total_volume = 0
    
    # 遍历路径上的每个时间点
    for position in path:
        # 计算当前位置的业务量并累加到总业务量上
        users_covered = calculate_users_covered(users, radius, position)
        total_volume += users_covered
    
    return total_volume

# 示例路径、用户位置和覆盖范围半径
# path = [(10, 10), (20, 20), (30, 30)]  # 示例路径，每个元素表示无人机的位置
# path= read_path_from_file('path/final_eva_path1_2024010_eva_.txt')

# 70
# users=[(30.15, 13.50), (60.63, 38.57), (64.12, 41.43), (79.40, 84.82), (16.00, 21.73), (18.80, 23.59), (90.30, 75.26), (73.53, 83.58), (72.15, 82.02), (80.44, 82.61), (56.41, 36.92), (76.78, 76.36), (26.85, 26.57), (26.66, 28.48), (76.94, 87.42), (20.93, 24.37), (16.98, 20.14), (20.30, 21.35), (15.21, 24.72), (19.87, 21.92), (14.75, 28.69), (74.96, 76.73), (28.35, 24.05), (54.68, 37.73), (83.64, 80.26), (59.69, 47.66), (83.12, 80.02), (55.92, 45.00), (76.58, 73.57), (16.36, 12.28), (79.15, 78.23), (61.91, 43.03), (24.25, 19.32), (53.33, 41.97), (84.78, 77.97), (59.89, 36.70), (69.67, 45.98), (66.68, 30.74), (78.62, 78.96), (57.93, 42.15), (20.69, 21.69), (12.78, 19.05), (78.03, 78.75), (20.57, 21.80), (18.20, 25.47), (81.23, 76.02), (59.56, 37.03), (76.91, 86.87), (24.24, 19.31), (28.42, 13.97), (16.11, 17.41), (59.92, 41.70), (8.58, 22.92), (53.63, 37.79), (28.35, 15.63), (12.37, 28.72), (81.28, 82.72), (79.17, 76.39), (82.59, 81.56), (81.25, 75.56), (52.92, 42.73), (68.75, 41.74), (54.52, 43.78), (82.98, 77.94), (56.67, 36.30), (68.11, 82.71), (12.71, 22.07), (59.41, 38.67), (58.61, 40.44)]
#30
# users=[(20.30, 21.35), (61.59, 43.16), (12.37, 28.72), (26.85, 26.57), (75.92, 85.00), (77.96, 80.08), (8.58, 22.92), (81.77, 75.11), (19.87, 21.92), (68.42, 33.97), (85.04, 75.17), (52.71, 42.07), (65.10, 34.48), (56.36, 32.28), (18.27, 18.65), (24.25, 19.32), (17.16, 18.38), (85.31, 81.42), (58.80, 43.59), (60.79, 38.20), (30.15, 13.50), (61.80, 36.94), (79.56, 77.03), (76.56, 82.41), (54.79, 44.03), (83.83, 77.09), (12.78, 19.05), (83.01, 81.82), (28.35, 15.63), (75.37, 83.05)]
# 39 users
# users=[(26.85, 26.57), (84.07, 81.42), (56.67, 36.30), (83.83, 77.09), (18.27, 18.65), (82.80, 72.84), (75.06, 80.23), (68.35, 44.05), (75.72, 78.45), (16.00, 21.73), (56.56, 42.41), (54.12, 37.94), (73.11, 80.28), (84.25, 80.39), (75.53, 73.09), (54.75, 48.69), (56.15, 48.54), (77.93, 82.15), (73.63, 77.79), (82.86, 72.77), (25.50, 25.38), (16.36, 12.28), (59.69, 47.66), (78.35, 78.93), (66.66, 48.48), (21.80, 16.94), (59.56, 37.03), (61.59, 43.16), (83.52, 77.91), (85.04, 75.17), (60.05, 47.99), (23.87, 16.68), (8.58, 22.92), (74.87, 78.04), (16.03, 11.14), (79.89, 76.70), (79.92, 81.70), (76.69, 86.90), (30.15, 13.50)]
# 50 user
users=[(77.57, 81.86), (20.20, 19.16), (26.85, 26.57), (17.72, 18.04), (56.67, 36.30), (20.69, 21.69), (23.87, 16.68), (56.15, 48.54), (84.07, 81.42), (18.61, 20.25), (63.45, 43.68), (88.75, 81.74), (79.39, 85.08), (18.27, 18.65), (16.99, 18.39), (59.69, 47.66), (55.53, 33.09), (69.67, 45.98), (86.68, 70.74), (54.22, 41.19), (19.87, 21.92), (62.34, 39.23), (23.40, 20.54), (28.35, 15.63), (74.16, 87.42), (61.68, 35.08), (81.78, 67.60), (16.11, 17.41), (70.92, 89.23), (76.76, 73.81), (80.63, 78.57), (77.33, 78.94), (59.82, 34.78), (80.44, 82.61), (15.31, 13.65), (55.06, 40.23), (62.58, 39.07), (90.30, 75.26), (17.72, 14.72), (15.21, 24.72), (61.33, 45.93), (77.93, 82.15), (20.05, 27.99), (54.87, 38.04), (81.28, 82.72), (65.31, 41.42), (79.92, 81.70), (21.40, 16.18), (63.01, 41.82), (27.89, 21.79), (74.68, 77.73)]		
# 60
# users=[(20.93, 24.37), (83.46, 74.71), (16.00, 21.73), (70.92, 89.23), (28.35, 24.05), (63.83, 37.09), (12.78, 19.05), (83.09, 83.11), (57.33, 38.94), (24.24, 19.31), (26.68, 19.08), (61.77, 35.11), (71.76, 77.86), (59.56, 37.03), (78.03, 78.75), (78.28, 80.42), (57.12, 30.06), (83.08, 85.82), (16.36, 12.28), (63.45, 43.68), (62.80, 32.84), (78.21, 81.65), (15.94, 20.23), (72.15, 82.02), (59.82, 34.78), (23.87, 16.68), (79.81, 78.72), (82.21, 79.48), (77.08, 78.25), (23.26, 20.58), (79.51, 84.56), (22.08, 13.13), (54.52, 43.78), (53.33, 41.97), (82.33, 77.42), (19.87, 21.92), (18.80, 23.59), (79.72, 79.52), (57.16, 43.03), (61.33, 45.93), (19.54, 21.84), (84.78, 77.97), (26.85, 26.57), (54.22, 41.19), (16.15, 28.54), (15.21, 24.72), (63.52, 37.91), (18.44, 20.70), (65.04, 35.17), (74.68, 77.73), (64.92, 42.93), (83.34, 74.68), (61.68, 35.08), (88.75, 81.74), (82.59, 81.56), (78.37, 80.70), (57.77, 48.63), (77.15, 78.22), (81.78, 67.60), (77.08, 77.67)]
# 70
# users=[(30.15, 13.50), (60.63, 38.57), (64.12, 41.43), (79.40, 84.82), (16.00, 21.73), (18.80, 23.59), (90.30, 75.26), (73.53, 83.58), (72.15, 82.02), (80.44, 82.61), (56.41, 36.92), (76.78, 76.36), (26.85, 26.57), (26.66, 28.48), (76.94, 87.42), (20.93, 24.37), (16.98, 20.14), (20.30, 21.35), (15.21, 24.72), (19.87, 21.92), (14.75, 28.69), (74.96, 76.73), (28.35, 24.05), (54.68, 37.73), (83.64, 80.26), (59.69, 47.66), (83.12, 80.02), (55.92, 45.00), (76.58, 73.57), (16.36, 12.28), (79.15, 78.23), (61.91, 43.03), (24.25, 19.32), (53.33, 41.97), (84.78, 77.97), (59.89, 36.70), (69.67, 45.98), (66.68, 30.74), (78.62, 78.96), (57.93, 42.15), (20.69, 21.69), (12.78, 19.05), (78.03, 78.75), (20.57, 21.80), (18.20, 25.47), (81.23, 76.02), (59.56, 37.03), (76.91, 86.87), (24.24, 19.31), (28.42, 13.97), (16.11, 17.41), (59.92, 41.70), (8.58, 22.92), (53.63, 37.79), (28.35, 15.63), (12.37, 28.72), (81.28, 82.72), (79.17, 76.39), (82.59, 81.56), (81.25, 75.56), (52.92, 42.73), (68.75, 41.74), (54.52, 43.78), (82.98, 77.94), (56.67, 36.30), (68.11, 82.71), (12.71, 22.07), (59.41, 38.67), (58.61, 40.44)]
# path= read_path_from_file('0_test_pathv1_9.txt')
path1=read_path_from_file('../best_path_1_2024097_eva.txt')
path2=read_path_from_file('../best_path_2_2024097_eva.txt')
path3=read_path_from_file('../best_path_3_2024097_eva.txt')

# path1=[(30, 15), (31, 15), (31, 14), (30, 14), (30, 13), (29, 13), (28, 13), (28, 12), (29, 12), (29, 11), (29, 10), (28, 10), (27, 10), (27, 9), (27, 8)]
# path2=[(50, 15), (49, 15), (49, 14), (48, 14), (48, 15), (47, 15), (47, 16), (48, 16), (49, 16), (50, 16), (50, 17), (50, 18), (49, 18), (48, 18), (48, 19)]
# path3=[(70, 15), (69, 15), (69, 14), (70, 14), (70, 13), (69, 13), (68, 13), (67, 13), (67, 12), (67, 11), (68, 11), (69, 11), (69, 12), (70, 12), (71, 12)]



radius = 10  
total_volume1 = total_business_volume(path1, users, radius)/len(path1)
total_volume1= total_volume1*100
total_volume2 = total_business_volume(path2, users, radius)/len(path2)
total_volume2= total_volume2*100
total_volume3 = total_business_volume(path3, users, radius)/len(path3)
total_volume3= total_volume3*100
print("总业务量:", total_volume1+total_volume2+total_volume3)
```
