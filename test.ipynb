```python
import numpy as np

def generate_cluster(center, covariance, num_users, x_bounds, y_bounds):
    users = []
    while len(users) < num_users:
        # 基于高斯分布生成一个用户
        user = np.random.multivariate_normal(center, covariance, 1)[0]
        # 如果用户坐标在指定范围内，则添加到用户列表中
        if x_bounds[0] <= user[0] <= x_bounds[1] and y_bounds[0] <= user[1] <= y_bounds[1]:
            users.append((user[0], user[1]))
    return users

clusters=[]
center_1 = (30, 15)
covariance_matrix = [[25, 0], [0, 25]]  # 假设x和y方向的标准差均为5
cluster = generate_cluster(center_1, covariance_matrix, 100, (0, 75), (-25, 25))
clusters.append(cluster)

center_2 = (50,0 )
covariance_matrix = [[25, 0], [0, 25]]  # 假设x和y方向的标准差均为5
cluster = generate_cluster(center_1, covariance_matrix, 100, (0, 75), (-25, 25))
clusters.append(cluster)

center_3 = (70, 15)
covariance_matrix = [[25, 0], [0, 25]]  # 假设x和y方向的标准差均为5
cluster = generate_cluster(center_1, covariance_matrix, 100, (0, 75), (-25, 25))
clusters.append(cluster)





print(clusters)  # 显示前5个用户的坐标
print(len(clusters))
print(len(clusters[0]))


import matplotlib.pyplot as plt

# 绘图
plt.figure(figsize=(10, 6))  # 创建一个图形实例，尺寸为10x6英寸
for cluster in clusters:
    # 解包每个群集中的x和y坐标
    x, y = zip(*cluster)
    plt.scatter(x, y)  # 为每个群集绘制散点图

# 设置图表标题和坐标轴标签
plt.title('plt')
plt.xlabel('X')
plt.ylabel('Y')

# 显示图表
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 创建一个网格
x, y = np.meshgrid(np.arange(100), np.arange(100))

# 设置高斯分布的均值和协方差矩阵
mu = np.array([50, 50])  # 均值
covariance = np.array([[100, 0], [0, 100]])  # 协方差矩阵

# 计算概率密度
x_diff = x - mu[0]
y_diff = y - mu[1]
inv_covariance = np.linalg.inv(covariance)
exponential_term = np.exp(-0.5 * (x_diff**2 * inv_covariance[0, 0] + 2 * x_diff * y_diff * inv_covariance[0, 1] + y_diff**2 * inv_covariance[1, 1]))
gaussian = (1.0 / (2 * np.pi * np.sqrt(np.linalg.det(covariance)))) * exponential_term

# 绘制高斯分布的等高线图
plt.contour(x, y, gaussian, levels=10, cmap='viridis')
plt.title('2D Gaussian Distribution')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Lyapunov function
def lyapunov(x, y):
    return x**2 + y**2

# Define the differential equations
def f(x, y):
    return -2*x

def g(x, y):
    return -2*y

# Define the grid
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

# Calculate Lyapunov function values
Z = lyapunov(X, Y)

# Plot Lyapunov function contour
plt.contour(X, Y, Z, levels=20)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lyapunov Function Contour')
plt.colorbar(label='Lyapunov Function Value')
plt.grid(True)
plt.show()

# Plot system trajectories
t = np.linspace(0, 10, 1000)
x0, y0 = 1, 1  # initial conditions
x_traj = np.zeros_like(t)
y_traj = np.zeros_like(t)
x_traj[0] = x0
y_traj[0] = y0
for i in range(1, len(t)):
    x_traj[i] = x_traj[i-1] + f(x_traj[i-1], y_traj[i-1]) * (t[i] - t[i-1])
    y_traj[i] = y_traj[i-1] + g(x_traj[i-1], y_traj[i-1]) * (t[i] - t[i-1])

# Plot system trajectory
plt.contour(X, Y, Z, levels=20)
plt.plot(x_traj, y_traj, label='System Trajectory', color='red')
plt.scatter(x_traj[0], y_traj[0], color='green', label='Initial Condition')
plt.xlabel('x')
plt.ylabel('y')
plt.title('System Trajectory with Lyapunov Function Contour')
plt.colorbar(label='Lyapunov Function Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def lawn_mower_path(rows, cols):
    path = []
    for i in range(rows):
        if i % 2 == 0:
            # Move from left to right
            for j in range(cols):
                path.append((i, j))
        else:
            # Move from right to left
            for j in range(cols-1, -1, -1):
                path.append((i, j))
    return path

def snake_path(rows, cols):
    path = []
    for i in range(rows):
        if i % 2 == 0:
            # Move from left to right
            for j in range(cols):
                path.append((i, j))
        else:
            # Move from right to left
            for j in range(cols-1, -1, -1):
                path.append((i, j))
    return path

def print_path(path, rows, cols):
    for i in range(rows):
        for j in range(cols):
            if (i, j) in path:
                print("X", end=" ")
            else:
                print(".", end=" ")
        print()

# Example usage
rows = 5
cols = 5

print("Lawn Mower Path:")
lawn_mower = lawn_mower_path(rows, cols)
print_path(lawn_mower, rows, cols)

print("\nSnake Path:")
snake = snake_path(rows, cols)
print_path(snake, rows, cols)
```
